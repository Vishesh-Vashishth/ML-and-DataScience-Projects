**Step 1:Import Libraries**


```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore") 
```

**Step2 : read Dataset**

```python
df = pd.read_csv("Downloads/Housing.csv");
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
df.mainroad.replace({'yes':1, 'no':0},inplace=True)
df.guestroom.replace({'yes':1, 'no':0},inplace=True)
df.basement .replace({'yes':1, 'no':0},inplace=True)
df.hotwaterheating.replace({'yes':1, 'no':0},inplace=True)
df.airconditioning  .replace({'yes':1, 'no':0},inplace=True)
df.prefarea .replace({'yes':1, 'no':0},inplace=True)
df.furnishingstatus.replace({'furnished':2,'semi-furnished':1, 'unfurnished':0},inplace=True)
```

```python
df.info()
```

```python
df.columns
```

```python
y = df['price']
X = df[['area', 'bedrooms', 'bathrooms', 'stories', 'mainroad',
       'guestroom', 'basement', 'hotwaterheating', 'airconditioning',
       'parking', 'prefarea','furnishingstatus']]
```

```python
import statsmodels.api as sm
print(sm.OLS(y,X).fit().summary())
```

**Train test split**

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y, random_state = 2529)
```

```python
import seaborn as sns 
```

```python
sns.pairplot(df)
```

```python
# we need to do some scaling before model selection for betterment of our result
from sklearn.preprocessing import StandardScaler 
ss = StandardScaler()
```

```python
X_train_ss = ss.fit_transform(X_train)
X_test_ss = ss.transform(X_test)
```

```python
pd.DataFrame(X_train_ss).describe()
```

**Step 5 : model selection**

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

**Step 6 : Train model**

```python
model.fit(X_train,y_train)
```

```python
model.intercept_
```

```python
model.coef_
```

**Step 7 : Prediction**

```python
y_pred = model.predict(X_test)
```

**Step 8 : Evaluation**

```python
from sklearn.metrics import mean_absolute_percentage_error
```

```python
mean_absolute_percentage_error(y_test,y_pred)
```
