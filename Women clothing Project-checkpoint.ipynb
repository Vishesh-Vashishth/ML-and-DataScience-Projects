## Project : Women clothings review and ratings

**Step 1: import libraries**

```python
import pandas as pd
import numpy as np
```

**Step 2: Import Dataset**

```python
df = pd.read_csv("https://github.com/YBIFoundation/ProjectHub/raw/main/Women%20Clothing%20E-Commerce%20Review.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
#we dropping title because title contain approx 4k na values and title has no work to do with model training so removing it will provivde more data to train our selected model
df.drop('Title',axis=1,inplace = True)
```

```python
df
```

```python
#now we will drop other na because we have enough values to train model
df.dropna(inplace= True)
```

**Step 3 : Define Y(output) and X(input)**

```python
y = df['Rating']
X = df['Review']
```

```python
X
```

**Step 4 Train-test split**

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=2529)
```

```python
X_train
```

```python
#Tokenisation
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(lowercase = True, analyzer='word', ngram_range=(1,2), stop_words = 'english',max_features=5000)
```

```python
X_train = cv.fit_transform(X_train)
```

```python
cv.get_feature_names_out()
```

```python
X_test = cv.transform(X_test)
```

```python
X_train.toarray().shape
```

**Step 5: Select Model**

```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
```

**Step 6 : Train model**

```python
model.fit(X_train,y_train)
```

**Step 7: Prediction**

```python
y_pred = model.predict(X_test)
```

**Step 8 : evalution**

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,y_pred))
```
